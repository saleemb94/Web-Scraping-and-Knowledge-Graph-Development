```python
from rdflib import Graph, Literal, Namespace, URIRef
from rdflib.namespace import RDF, RDFS
```

```python
from rdflib import Graph, Literal, Namespace, URIRef
from rdflib.namespace import RDF, RDFS,OWL


psut_onto = Namespace("http://example.com/PSUT_Onto/")
owl = Namespace("http://www.w3.org/2002/07/owl#")
rdf = Namespace("http://www.w3.org/1999/02/22-rdf-syntax-ns#")
xml = Namespace("http://www.w3.org/XML/1998/namespace")
xsd = Namespace("http://www.w3.org/2001/XMLSchema#")
rdfs = Namespace("http://www.w3.org/2000/01/rdf-schema#")


g = Graph()
g.bind("ex", psut_onto)
```

```python



# Process school data
for school_entry in school_data:
    School_uri = psut_onto[school_entry["School Name"].replace(" ", "_")]

    # Add triples for school
    g.add((School_uri, RDF.type, psut_onto.School))
    g.add((School_uri, rdfs.label, Literal(school_entry["School Name"])))
    g.add((School_uri, psut_onto.has_dean, Literal(school_entry["Dean"])))
    g.add((School_uri, psut_onto.has_url, URIRef(school_entry["URL"])))

    #print(school_entry)

    # Add triples for departments
    if "Departments" in school_entry:
      for department in school_entry["Departments"]:

          g.add((School_uri, psut_onto.has_department, Literal(department)))

    if "Research Departments" in school_entry:
      for Rdepartment in school_entry["Research Departments"]:
          print(Rdepartment)
          g.add((School_uri, psut_onto.has_research_department, Literal(Rdepartment)))

    # Add triples for programs
    for program in school_entry["Prgorams"]:
        print(program)
        g.add((School_uri, psut_onto.has_program, Literal(program)))

# Process staff data
for staff_entry in staff_data:
    staff_uri = psut_onto[staff_entry["Name"].replace(" ", "_").replace('"', "").replace("'", "").replace("\\","")]
    school_uri = URIRef(psut_onto + staff_entry["School"].replace(" ", "_"))

    # Add triples for staff
    g.add((staff_uri, RDF.type, psut_onto.Staff))
    g.add((staff_uri, rdfs.label, Literal(staff_entry["Name"].replace(" ", "_").replace('"', "").replace("'", "").replace("\\",""))))
    g.add((staff_uri, psut_onto.has_email, Literal(staff_entry["Email"])))
    g.add((staff_uri, psut_onto.has_position1, Literal(staff_entry["Position_1"])))
    g.add((staff_uri, psut_onto.has_position2, Literal(staff_entry["Position_2"])))
    g.add((staff_uri, psut_onto.has_summary, Literal(staff_entry["Summary"])))
    g.add((staff_uri, psut_onto.has_telephone, Literal(staff_entry["Telephone and EXT"])))

    g.add((staff_uri, psut_onto.WorksIn, school_uri))
```

```python
out = g.serialize(format='turtle')
```

```python
# Query 1: All Programs of one school at PSUT


PREFIX ex: <http://example.com/PSUT_Onto/>
PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
SELECT ?school ?program
WHERE {
  ?school a ex:School ;
  rdfs:label "King Talal School of Business Technology";
  ex:has_program ?program .
}
```

```python
# Query 2: All departments of one school at PSUT


PREFIX ex: <http://example.com/PSUT_Onto/>
PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
SELECT ?school ?department
WHERE {
  ?school a ex:School ;
  rdfs:label "King Talal School of Business Technology";
  ex:has_department ?department .
}
```

```python
# Query 3: All deans names of PSUT schools

PREFIX ex: <http://example.com/PSUT_Onto/>
PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
SELECT ?school ?dean
WHERE {
  ?school a ex:School ;
    rdfs:label ?schoolname;

    ex:has_dean ?dean .
}
```

```python
# Query 4: Get the name and Email of the president and vice president of PSUT

PREFIX ex: <http://example.com/PSUT_Onto/>
PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
SELECT ?staff ?position ?Email ?schoolName
WHERE {
  ?staff a ex:Staff ;

    ex:has_position2 ?position ;
   	ex:has_email ?Email;
    ex:WorksIn ?school .
    ?school rdfs:label ?schoolName .
     FILTER (CONTAINS(?position, "President"))

}
```

```python
#Query 5: Get the names, Email and numbers of all deans and vice deans of PSUT

PREFIX ex: <http://example.com/PSUT_Onto/>
PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
SELECT ?staff ?position ?schoolName ?Email ?Number
WHERE {
  ?staff a ex:Staff ;

    ex:has_position2 ?position ;
   	ex:has_email ?Email;
    ex:has_telephone ?Number;
    ex:WorksIn ?school .
    ?school rdfs:label ?schoolName .
     FILTER (CONTAINS(?position, "Dean"))

}
```

```python
#Query 6: Get the names and contact info of professors in one school of PSUT

PREFIX ex: <http://example.com/PSUT_Onto/>
PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
SELECT ?staff ?position ?schoolName ?Email ?Number
WHERE {
  ?staff a ex:Staff ;

    ex:has_position1 ?position ;

   	ex:has_email ?Email;
    ex:has_telephone ?Number;
    ex:WorksIn ex:King_Abdullah_II_School_of_Engineering .
    ex:King_Abdullah_II_School_of_Engineering rdfs:label ?schoolName .
     FILTER (CONTAINS(UCASE(?position), "PROFESSOR"))

}
```

```python
# Qurey 7: Get the summary of one of the deans of PSUT schools

PREFIX ex: <http://example.com/PSUT_Onto/>
PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
SELECT ?staff ?position ?Summary
WHERE {
  ?staff a ex:Staff ;

    ex:has_position2 ?position ;

   	ex:has_summary ?Summary;
    ex:WorksIn ex:King_Abdullah_II_School_of_Engineering .
    ex:King_Abdullah_II_School_of_Engineering rdfs:label ?schoolName .
     FILTER (?position= "Dean of King Abdullah II School of Engineering")

}
```

```python
#Query 8: Get the URLs of each school

PREFIX ex: <http://example.com/PSUT_Onto/>
PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
SELECT ?school ?schoolname ?URL
WHERE {
  ?school a ex:School ;
  rdfs:label ?schoolname;
    ex:has_url ?URL .



}
```
